<h3>Printing data to look at the columns required for preprocessing</h3>

```python

import pandas as pd
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import operator
from collections import defaultdict
import math
%matplotlib inline
```

```python
#reading the file
df = pd.read_csv("viz1.csv")
df.head(10)
```

```python
df.columns
```

<h3>Looping through school years as keys for the data file required for the visualizations</h3>

```python
years = Counter(df["School Year"])
school_years = []
for key in years:
    school_years.append(key)
```

```python
school_years
```

<h3>Populating the required data in nested dictionaries</h3>

```python
enroll = defaultdict(lambda : defaultdict(lambda: defaultdict(list)))

#        'Enroll, Female: American Indian/Alaska Native (Enrl F)',
#        'Enroll, Female: Asian (Enrl F)',
#        'Enroll, Female: Black/African American (Enrl F)',
#        'Enroll, Female: Hispanics of any race (Enrl F)',
#        'Enroll, Female: Native Hawaiian/Other Pacific Islander (Enrl F)',
#        'Enroll, Female: Two or more races (Enrl F)',
#        'Race/Ethnicity Unknown (Enrl F)', 'Enroll, Female: White (Enrl F)',
#        'Enroll, Male: American Indian/Alaska Native (Enrl M)',
#        'Enroll, Male: Asian (Enrl M)',
#        'Enroll, Male: Black/African American (Enrl M)',
#        'Enroll, Male: Hispanics of any race (Enrl M)',
#        'Enroll, Male: Native Hawaiian/Other Pacific Islander (Enrl M)',
#        'Enroll, Male: Two or more races (Enrl M)',
#        'Race/Ethnicity Unknown (Enrl M)', 'Enroll, Male: White (Enrl M)',

for i,row in df.iterrows():
    if row["School Year"] in school_years:
        #American Indian/Alaska Native 
        #Female
        if not math.isnan(row["Enroll, Female: American Indian/Alaska Native (Enrl F)"]):
            enroll[str(row["School Year"])]["female"]["American Indian/Alaska Native"].append(row["Enroll, Female: American Indian/Alaska Native (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["American Indian/Alaska Native"].append(row["Enroll, Female: American Indian/Alaska Native (Enrl F)"])
            enroll[str(row["School Year"])]["female"]["all"].append(row["Enroll, Female: American Indian/Alaska Native (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Female: American Indian/Alaska Native (Enrl F)"])
        else:
            pass
        
        #Male
        if not math.isnan(row["Enroll, Male: American Indian/Alaska Native (Enrl M)"]):
            enroll[str(row["School Year"])]["male"]["American Indian/Alaska Native"].append(row["Enroll, Male: American Indian/Alaska Native (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["American Indian/Alaska Native"].append(row["Enroll, Male: American Indian/Alaska Native (Enrl M)"])
            enroll[str(row["School Year"])]["male"]["all"].append(row["Enroll, Male: American Indian/Alaska Native (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Male: American Indian/Alaska Native (Enrl M)"])
        else:
            pass
        
        #Asian
        #Female
        if not math.isnan(row["Enroll, Female: Asian (Enrl F)"]):
            enroll[str(row["School Year"])]["female"]["Asian"].append(row["Enroll, Female: Asian (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["Asian"].append(row["Enroll, Female: Asian (Enrl F)"])
            enroll[str(row["School Year"])]["female"]["all"].append(row["Enroll, Female: Asian (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Female: Asian (Enrl F)"])
        else:
            pass

        #Male
        if not math.isnan(row["Enroll, Male: Asian (Enrl M)"]):
            enroll[str(row["School Year"])]["male"]["Asian"].append(row["Enroll, Male: Asian (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["Asian"].append(row["Enroll, Male: Asian (Enrl M)"])
            enroll[str(row["School Year"])]["male"]["all"].append(row["Enroll, Male: Asian (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Male: Asian (Enrl M)"])
        else:
            pass
        
        #Black/African American
        #Female
        if not math.isnan(row["Enroll, Female: Black/African American (Enrl F)"]):
            enroll[str(row["School Year"])]["female"]["Black/African American"].append(row["Enroll, Female: Black/African American (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["Black/African American"].append(row["Enroll, Female: Black/African American (Enrl F)"])
            enroll[str(row["School Year"])]["female"]["all"].append(row["Enroll, Female: Black/African American (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Female: Black/African American (Enrl F)"])
        else:
            pass

        #Male
        if not math.isnan(row["Enroll, Male: Black/African American (Enrl M)"]):
            enroll[str(row["School Year"])]["male"]["Black/African American"].append(row["Enroll, Male: Black/African American (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["Black/African American"].append(row["Enroll, Male: Black/African American (Enrl M)"])
            enroll[str(row["School Year"])]["male"]["all"].append(row["Enroll, Male: Black/African American (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Male: Black/African American (Enrl M)"])
        else:
            pass
        
        #Hispanic
        #Female
        if not math.isnan(row["Enroll, Female: Hispanics of any race (Enrl F)"]):
            enroll[str(row["School Year"])]["female"]["Hispanic"].append(row["Enroll, Female: Hispanics of any race (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["Hispanic"].append(row["Enroll, Female: Hispanics of any race (Enrl F)"])
            enroll[str(row["School Year"])]["female"]["all"].append(row["Enroll, Female: Hispanics of any race (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Female: Hispanics of any race (Enrl F)"])
        else:
            pass

        #Male
        if not math.isnan(row["Enroll, Male: Hispanics of any race (Enrl M)"]):
            enroll[str(row["School Year"])]["male"]["Hispanic"].append(row["Enroll, Male: Hispanics of any race (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["Hispanic"].append(row["Enroll, Male: Hispanics of any race (Enrl M)"])
            enroll[str(row["School Year"])]["male"]["all"].append(row["Enroll, Male: Hispanics of any race (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Male: Hispanics of any race (Enrl M)"])
        else:
            pass
        
        #Native Hawaiian/Other Pacific Islander
        #Female
        if not math.isnan(row["Enroll, Female: Native Hawaiian/Other Pacific Islander (Enrl F)"]):
            enroll[str(row["School Year"])]["female"]["Native Hawaiian/Other Pacific Islander"].append(row["Enroll, Female: Native Hawaiian/Other Pacific Islander (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["Native Hawaiian/Other Pacific Islander"].append(row["Enroll, Female: Native Hawaiian/Other Pacific Islander (Enrl F)"])
            enroll[str(row["School Year"])]["female"]["all"].append(row["Enroll, Female: Native Hawaiian/Other Pacific Islander (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Female: Native Hawaiian/Other Pacific Islander (Enrl F)"])
        else:
            pass

        #Male
        if not math.isnan(row["Enroll, Male: Native Hawaiian/Other Pacific Islander (Enrl M)"]):
            enroll[str(row["School Year"])]["male"]["Native Hawaiian/Other Pacific Islander"].append(row["Enroll, Male: Native Hawaiian/Other Pacific Islander (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["Native Hawaiian/Other Pacific Islander"].append(row["Enroll, Male: Native Hawaiian/Other Pacific Islander (Enrl M)"])
            enroll[str(row["School Year"])]["male"]["all"].append(row["Enroll, Male: Native Hawaiian/Other Pacific Islander (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Male: Native Hawaiian/Other Pacific Islander (Enrl M)"])
        else:
            pass
        
        #White
        #Female
        if not math.isnan(row["Enroll, Female: White (Enrl F)"]):
            enroll[str(row["School Year"])]["female"]["White"].append(row["Enroll, Female: White (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["White"].append(row["Enroll, Female: White (Enrl F)"])
            enroll[str(row["School Year"])]["female"]["all"].append(row["Enroll, Female: White (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Female: White (Enrl F)"])
        else:
            pass

        #Male
        if not math.isnan(row["Enroll, Male: White (Enrl M)"]):
            enroll[str(row["School Year"])]["male"]["White"].append(row["Enroll, Male: White (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["White"].append(row["Enroll, Male: White (Enrl M)"])
            enroll[str(row["School Year"])]["male"]["all"].append(row["Enroll, Male: White (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Male: White (Enrl M)"])
        else:
            pass
        
        #Two or more races
        #Female
        if not math.isnan(row["Enroll, Female: Two or more races (Enrl F)"]):
            enroll[str(row["School Year"])]["female"]["Two or more races"].append(row["Enroll, Female: Two or more races (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["Two or more races"].append(row["Enroll, Female: Two or more races (Enrl F)"])
            enroll[str(row["School Year"])]["female"]["all"].append(row["Enroll, Female: Two or more races (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Female: Two or more races (Enrl F)"])
        else:
            pass

        #Male
        if not math.isnan(row["Enroll, Male: Two or more races (Enrl M)"]):
            enroll[str(row["School Year"])]["male"]["Two or more races"].append(row["Enroll, Male: Two or more races (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["Two or more races"].append(row["Enroll, Male: Two or more races (Enrl M)"])
            enroll[str(row["School Year"])]["male"]["all"].append(row["Enroll, Male: Two or more races (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Enroll, Male: Two or more races (Enrl M)"])
        else:
            pass
        
        #Race/Ethnicity Unknown
        #Female
        if not math.isnan(row["Race/Ethnicity Unknown (Enrl F)"]):
            enroll[str(row["School Year"])]["female"]["Race/Ethnicity Unknown"].append(row["Race/Ethnicity Unknown (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["Race/Ethnicity Unknown"].append(row["Race/Ethnicity Unknown (Enrl F)"])
            enroll[str(row["School Year"])]["female"]["all"].append(row["Race/Ethnicity Unknown (Enrl F)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Race/Ethnicity Unknown (Enrl F)"])
        else:
            pass

        #Male
        if not math.isnan(row["Race/Ethnicity Unknown (Enrl M)"]):
            enroll[str(row["School Year"])]["male"]["Race/Ethnicity Unknown"].append(row["Race/Ethnicity Unknown (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["Race/Ethnicity Unknown"].append(row["Race/Ethnicity Unknown (Enrl M)"])
            enroll[str(row["School Year"])]["male"]["all"].append(row["Race/Ethnicity Unknown (Enrl M)"])
            enroll[str(row["School Year"])]["all"]["all"].append(row["Race/Ethnicity Unknown (Enrl M)"])
        else:
            pass
#enroll
```
